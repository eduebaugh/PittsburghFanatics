## How the Data is Used

The data will be used to describe what the best neighborhood in pittsburgh is by showing which neighborhood has the least amount of crashes, as well as what were the causes of the crashes.

## Metrics

* Municipalities with lowest frequency in data sheet
* Municipalities with lowest number of drug and alcohol related accidents
* Street condition at crash sites

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

bikes = pd.read_csv('bikes.csv')
bikes.head(10)
```

```python
st = 'Street'
bld = 'Bike Lane Distance (Miles)'
sld = 'Shared Lane Marking Distance (Miles)'

bikes = bikes[[st, bld, sld]]
bikes.head(10)
```

```python
bld_avg = bikes[bld].mean()
bld_df = bikes[bikes[bld] > bld_avg].sort_values(by=bld, ascending=False)
bld_df = bld_df[[st, bld]]

bld_df.head(10)
```

```python
bld_df.plot.bar(x=st, y=bld)
```

```python
sld_avg = bikes[sld].mean()
sld_df = bikes[bikes[sld] > sld_avg].sort_values(by=sld, ascending=False)
sld_df = sld_df[[st, sld]]

sld_df.head(10)
```

```python
sld_df.plot.bar(x=st, y=sld)
```

```python
cmb = 'Combined Distance (Miles)'
for index, row in bikes.iterrows():
    x = row[bld]
    y = row[sld]
    if np.isnan(x):
        x = 0
    if np.isnan(y):
        y = 0
    bikes.at[index, cmb] = x + y

bikes.head(10)
```

```python
cmb_avg = bikes[cmb].mean()
cmb_df = bikes[bikes[cmb] > cmb_avg].sort_values(by=cmb, ascending=False)
cmb_df = cmb_df[[st, cmb]]

cmb_df.head(10)
```

```python
cmb_df.plot.bar(x=st, y=cmb)
```

```python
st_to_nbh = {
    'Beechwood Ave' : 'Squirrel Hill',
    'Forbes Ave' : 'North Oakland',
    'E Carson St' : 'South Side Flats',
    'Butler St' : 'Central Lawrenceville',
    'Liberty Ave' : 'Bloomfield',
    'Brighton Rd' : 'Brighton Heights',
    'O\'Hara Bigelow Bayard' : 'North Oakland',
    'Riverview Dr' : 'Brighton Heights',
    'Highland Ave' : 'Highland Park',
    'East Liberty Blvd' : 'East Liberty',
    'Negley Run' : 'Highland Park',
    'Penn Ave' : 'East Liberty',
    'Morewood Ave' : 'Shadyside',
    'Dallas Ave' : 'Point Breeze',
    'Ellsworth Ave' : 'Shadyside',
    'Schenley Dr' : 'Squirrel Hill',
    'Friendship Ave' : 'Friendship',
    'Hamilton Ave' : 'Larimer',
    'Braddock Ave' : 'Point Breeze',
    'Ellsworth' : 'Shadyside',
    'Birmingham Bridge' : 'South Side Flats',
    'Reynolds' : 'Point Breeze',
}

d = dict()
for index, row in cmb_df.iterrows():
    nbh = st_to_nbh[row[st]]
    dist = row[cmb]
    if nbh not in d:
        d[nbh] = dist
    else:
        d[nbh] += dist
        
nbh = 'Neighborhood'
tot = 'Total Distance (Miles)'
keys = list(d.keys())
values = list(d.values())
data = {
    nbh : keys,
    tot : values
}
neighborhoods = pd.DataFrame.from_dict(data)

neighborhoods.head(10)
```

```python
neighborhoods.plot.bar(x=nbh, y=tot)
```
